```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
from scipy import stats
```

```python
#1.1 Create Random 4x4 Array from Exponential Distribution
# Create a 4x4 array from exponential distribution
# scale parameter controls the shape of the distribution
exponential_4x4 = np.random.exponential(scale=1.0, size=(4, 4))
print("4x4 Exponential Distribution Array:")
print(exponential_4x4)
```

```python
plt.hist(np.random.rand(100000), density=True, bins=100, histtype="step", color="blue",
label="rand")
plt.axis([-2.5, 2.5, 0, 1.1])
plt.legend(loc = "upper left")
plt.title("Random distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
# Create random arrays from different distributions
exponential_100k = np.random.exponential(scale=1.0, size=100000)
uniform_100k = np.random.rand(100000)
normal_100k = np.random.normal(0, 1, 100000)

# Create the histogram plot
plt.figure(figsize=(10, 6))

# Plot exponential distribution
plt.hist(exponential_100k, density=True, bins=100, histtype="step",
         color="red", label="Exponential", alpha=0.7)

# Plot uniform distribution
plt.hist(uniform_100k, density=True, bins=100, histtype="step",
         color="blue", label="Uniform", alpha=0.7)

# Plot normal distribution
plt.hist(normal_100k, density=True, bins=100, histtype="step",
         color="green", label="Normal", alpha=0.7)

# Adjust axis for better visualization
plt.axis([-3, 6, 0, 2.5])
plt.legend(loc="upper right")
plt.title("Comparison of Random Distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Create meshgrid for x and y values
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# Calculate Z = X² + Y²
Z = X**2 + Y**2

# Create 3D plot
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# Surface plot
surf = ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.8)

# Add labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z = X² + Y²')
ax.set_title('3D Plot: Z = X² + Y²')

# Add colorbar
fig.colorbar(surf, ax=ax, shrink=0.5, aspect=5)

plt.show()
```
